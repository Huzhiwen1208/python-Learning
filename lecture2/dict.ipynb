````python
wtk = {
    "name": "WTK",
    "age": 12,
    "score": 100,
    "height": 170.2
}
# 增查删改
##  增
wtk["weight"] = 90
print(wtk)

## 取年龄
print(wtk["age"])
print(wtk.get("age"))
print(wtk.get("math"))

## 读取字典的范式
if wtk.get("age") is not None:
    print(wtk["age"])

## 删除字典某个信息的范式
if wtk.get("math") is not None:
    del wtk["math"]
print(wtk)

# 修改: 如果key存在，则修改，否则创建新的key-value
wtk["score"] = 99
print(wtk)

wtk["math"] = 88
print(wtk)
````
