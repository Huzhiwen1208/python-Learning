# 字符串
CRUD：增查删改  Create、Read、Update、Delete
英语字母、中文汉字、日语字 ==> 字符
多个字符连起来就是字符串

````python
# Create
name = "Tom"
print(name)
# 增加
name += " Jimmy"
print(name)

# 读/查 J 字符: 正着数/倒着数
print(name[4])
print(name[-5])

# 读/查 Jim 字符串: 正着数/倒着数
# J: 4/-5  m: 6/-3 m的后一个位置: 7/-2
print(name[4:7])
print(name[4:-2])
print(name[-5:7])
print(name[-5:-2])

# 取 miJ: 不能从后往前取：先取 Jim，再 reverse
jim = name[4:7]
# reverse: 如何通过冒号进行字符串的翻转 TODO
print(jim[-1:0])

# 删除
print(name)
# 不支持 del name[4]
# 1. 通过转列表，删除列表的中的 item，再转回来, @TODO 学弟完成
# 2. 通过切片再组合方式
print(name[:4]+name[5:])
````

````python
name = "Tom Jimmy"
print("find index of substr: ", name.find("Jim"))
str1 = "ABC0011.."
str2 = "ABCDEFGasdjlakjk"
print(str1.isalpha())
print(str2.isalpha())

name = name.replace("Jim", "XXX")
print(name)
````
